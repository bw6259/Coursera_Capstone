```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
html = urllib.request.urlopen(url).read()
#print(html)
postcode = []
borough = []
neighbourhood = []

soup = BeautifulSoup(html, 'html5lib') #lxml html5lib
table = soup.find('table',class_="wikitable") # sortable jquery-tablesorter
#print(table)
test = table.tbody #tr.th.
a=0
b=1
c=2
for rows in enumerate(test.find_all('td')):
    #print(rows)
    if rows[0] == a:
        a = a+3
        postcode.append(rows[1].text)
    elif rows[0] == b:
        b = b+3
        borough.append(rows[1].text)
    elif rows[0] == c:
        c = c+3
        neighbourhood.append(rows[1].text[:-1])
#print(postcode)    
#print(borough)
#print(neighbourhood)   
```

```python
wikipage_dict = {'Postcode':postcode, 'Borough':borough, 'Neighbourhood':neighbourhood}
wikipage_df = pd.DataFrame(wikipage_dict)
wikipage_df.set_index('Postcode', inplace=True)
wikipage_df
#print(postcode)    
#print(borough)
#print(neighbourhood) 
```

```python
wikipage_df['Borough'].value_counts()
```

```python
wiki_notassigned = wikipage_df[wikipage_df.Borough != 'Not assigned']
wiki_notassigned
```

```python
wiki_notassigned = wiki_notassigned.reset_index()
#wiki_grouped = wiki_notassigned.groupby(['Postcode','Borough'])
wiki_grouped = wiki_notassigned.groupby('Postcode', as_index=False).agg({'Neighbourhood' : ', '.join, 'Borough' : 'first'})
wiki_grouped
```

```python
wiki_grouped = wiki_grouped[['Postcode','Borough','Neighbourhood']]
wiki_grouped
```

```python
wiki_clean = wiki_grouped
for item in wiki_clean['Neighbourhood']:
    if item == 'Not assigned':
        item == wiki_clean['Borough']
wiki_clean
```
